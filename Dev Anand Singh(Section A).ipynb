**Q1. What is the difference between the `is` operator and the `==` operator? Give one example where two variables return True for `==` but False for `is`.**

**Answer:**
`==` (equality operator) checks whether two objects have the same value.
`is` (identity operator) checks whether two variables refer to the exact same object in memory.

Example where `==` is True but `is` is False:
```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True  — same values
print(a is b)  # False — different objects in memory
```
a == b is True because both lists contain the same elements.

a is b is False because a and b are two different list objects stored at different memory locations.

````python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)
print(a is b)
````

**Q3. Predict the output of `print(0.1 + 0.2 == 0.3)`. Then explain WHY this happens. (Hint: it is not a
Python bug.)**

**Answer:**
`0.1 + 0.2 == 0.3` → `False`**

It is a fundamental limitation of floating-point representation. Computers store numbers in binary, and fractions like 0.1 and 0.2 cannot be represented exactly in base-2. The stored approximations add up to something like 0.30000000000000004, which is not exactly equal to 0.3.

**Q5. Why is dictionary lookup O(1) on average while searching for a value in a list using `in` is O(n)?
Explain in plain English without using the word "hash" - then explain it again WITH the word "hash".**

**Answer:**

**Without using the word "hash"**

Dictionary lookup – O(1) on average: A dictionary uses a special method to determine where a key should be stored, so it can go directly to that location instead of checking every item.

List search using in – O(n): A list stores items one after another. To find a value, Python may have to check each element from the beginning until it finds a match or reaches the end.

**With the word "hash"**

Dictionary lookup – O(1) on average:A dictionary computes a hash of the key and uses the hash value to locate the correct position where the key-value pair is stored. This allows Python to access the item directly, making lookup O(1) on average.

List search using in – O(n):A list has no hash-based indexing. Python must compare the target value with each element one by one until it finds a match or reaches the end, resulting in O(n) time complexity.

**Q10. Why is NumPy significantly faster than Python lists for numerical operations? Give 3 reasons, atleast one should mention memory layout**

**Answer:**

NumPy is much faster than Python lists for numerical operations because:

**1.Contiguous memory layout:** NumPy arrays store elements in contiguous (continuous) memory, improving cache performance and making data access much faster. Python lists store references to objects scattered in memory.

**2.Vectorized operations:** NumPy performs operations on the entire array at once, avoiding slow Python loops.

**3.Fixed data type:** All elements in a NumPy array have the same data type, so NumPy does not need to check the type of each element during computations. Python lists can contain mixed data types, which adds overhead.
