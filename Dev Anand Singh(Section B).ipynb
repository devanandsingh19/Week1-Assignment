**Q11. Predict the output. Why does the second line behave differently from the first?**

```python
matrix1 = [[0] * 3 for _ in range(3)]
matrix2 = [[0] * 3] * 3

matrix1[0][0] = 99
matrix2[0][0] = 99

print(matrix1)
print(matrix2)
```
**Answer:**

**Output:**

```
[[99, 0, 0], [0, 0, 0], [0, 0, 0]]
[[99, 0, 0], [99, 0, 0], [99, 0, 0]]
```
**Reason:**

- `matrix1 = [[0] * 3 for _ in range(3)]` The list comprehension creates three separate inner lists. Changing one row does not affect the others.
- `matrix2 = [[0] * 3] * 3` [[0] * 3] creates one inner list. `* 3` does not create new lists, it creates three references to the same inner list.

````python
matrix1 = [[0] * 3 for _ in range(3)]
matrix2 = [[0] * 3] * 3
matrix1[0][0] = 99
matrix2[0][0] = 99
print(matrix1)
print(matrix2)
````

**Q12. Predict the output:**
```
x = [1, 2, 3]
y = x
x = x + [4]
print("x =", x)
print("y =", y)
a = [1, 2, 3]
b = a
a += [4]
print("a =", a)
print("b =", b)
```
**Now answer: Why do `x = x + [4]` and `a += [4]` behave differently?**

**Answer:**

**Output:**

```
x = [1, 2, 3, 4]
y = [1, 2, 3]
a = [1, 2, 3, 4]
b = [1, 2, 3, 4]
```

- `x = x + [4]` x + [4] creates a new list. The variable x is updated to reference this new list. y still refers to the original list, so it remains unchanged.
- `a += [4]` += modifies the existing list in place. Since a and b refer to the same list, both variables see the change.

````python
x = [1, 2, 3]
y = x
x = x + [4]
print("x =", x)
print("y =", y)
a = [1, 2, 3]
b = a
a += [4]
print("a =", a)
print("b =", b)
````

```python
for i in range(5):
 pass

for j in range(0):
 pass

```

**Answer:**

**Output:**

```python
4

```

**Reason:**

**Q14. Predict the output of each print statement. Then state which lines will raise an error (if any).**
```
print("10" + "20")
print(10 + 20)
print("10" * 3)
print(10 * "3")
print("10" + 20)
print(int("10") + 20)
```
**Answer:**

| Statement               | Output        |
| ----------------------- | ------------- |
| `print("10" + "20")`    | `1020`        |
| `print(10 + 20)`        | `30`          |
| `print("10" * 3)`       | `101010`      |
| `print(10 * "3")`       | `3333333333`  |
| `print("10" + 20)`      | **TypeError** |
| `print(int("10") + 20)` | `30`          |

**Reason:**

- `"10" + "20"` → Concatenates two strings → 1020
- `10 + 20` → Adds two integers → 30
- `"10" * 3` → Repeats the string 3 times → 101010
- `10 * "3"` → Also repeats the string 10 times → 3333333333
- `"10" + 20` → Raises TypeError because Python cannot concatenate a string and an integer.
- `int("10") + 20` → Converts "10" to an integer, then performs addition → 30

````python
print("10" + "20")
print(10 + 20)
print("10" * 3)
print(10 * "3")
print("10" + 20)
````

**Q15. Predict the output. Then explain Python's short-circuit evaluation in one line.**
```
print(0 or 5)
print(5 or 0)
print(None or "hello")
print(0 and 5)
print(5 and 10)
print([] or "default")
print([0] or "default") # careful!
```

**Answer:**

| Statement                 | Output    |
| ------------------------- | --------- |
| `print(0 or 5)`           | `5`       |
| `print(5 or 0)`           | `5`       |
| `print(None or "hello")`  | `hello`   |
| `print(0 and 5)`          | `0`       |
| `print(5 and 10)`         | `10`      |
| `print([] or "default")`  | `default` |
| `print([0] or "default")` | `[0]`     |

**Reason:**

- `0 or 5` → 0 is False, so or returns 5.
- `5 or 0` → 5 is True, so or returns 5.
- `None or "hello"` → None is False, so or returns "hello".
- `0 and 5` → 0 is False, so and returns 0 immediately.
- `5 and 10` → 5 is True, so and returns the second operand (10).
- `[] or "default"` → An empty list is False, so or returns "default".
- `[0] or "default"` → [0] is a non-empty list, which is True, so or returns [0].

````python
print(0 or 5)
print(5 or 0)
print(None or "hello")
print(0 and 5)
print(5 and 10)
print([] or "default")
print([0] or "default")
````
