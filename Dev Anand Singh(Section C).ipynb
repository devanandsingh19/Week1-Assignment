**Q31. Find all Armstrong numbers between 1 and 9999. (An n-digit number is Armstrong if the sum of
its digits each raised to the power n equals the number itself. Example: 153 = 1³ + 5³ + 3³.) Print them
in a clean comma-separated list with a count at the end.**

````python
armstrong_numbers = []

for num in range(1, 10000):
    n_digits = len(str(num))
    temp = num
    digit_power_sum = 0
    while temp > 0:
        digit = temp % 10
        digit_power_sum += digit ** n_digits
        temp //= 10
    if digit_power_sum == num:
        armstrong_numbers.append(num)

print("Armstrong numbers:", ", ".join(str(n) for n in armstrong_numbers))
print(f"Count: {len(armstrong_numbers)}")
````

**Q33. Prime check, optimized. Take a number N. Check if it is prime by iterating only up to √N. Then,
print the SUM of all primes between 2 and N (inclusive). Do not use any library.**

````python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:   # only check up to √n
        if n % i == 0:
            return False
        i += 2
    return True

N = int(input("Enter N: "))
print(f"{N} is {'prime' if is_prime(N) else 'not prime'}")

prime_sum = sum(i for i in range(2, N + 1) if is_prime(i))
print(f"Sum of all primes from 2 to {N}: {prime_sum}")
````

````python
review = input("Enter a Swiggy review: ")
words = review.lower().split()

freq = {}
for word in words:
    clean_word = ""
    for ch in word:
        if ch.isalpha():
            clean_word += ch
    if clean_word:
        freq[clean_word] = freq.get(clean_word, 0) + 1

items = list(freq.items())  
for i in range(len(items)):
    for j in range(i + 1, len(items)):
        if items[j][1] > items[i][1]:
            items[i], items[j] = items[j], items[i]

print("\nWord Frequencies (descending):")
for word, count in items:
    print(f"  {word}: {count}")

print(f"\nMost frequent: '{items[0][0]}' ({items[0][1]} times)")
````

````python
prices = (3845, 3902, 3878, 3955, 4012, 3998, 4055)
days = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")

max_price = prices[0]
max_day = days[0]
for i in range(len(prices)):
    if prices[i] > max_price:
        max_price = prices[i]
        max_day = days[i]

min_price = prices[0]
min_day = days[0]
for i in range(len(prices)):
    if prices[i] < min_price:
        min_price = prices[i]
        min_day = days[i]

avg = sum(prices) / len(prices)

max_gain = prices[1] - prices[0]
gain_from = days[0]
gain_to = days[1]
for i in range(1, len(prices)):
    gain = prices[i] - prices[i - 1]
    if gain > max_gain:
        max_gain = gain
        gain_from = days[i - 1]
        gain_to = days[i]


volatility = 0
for i in range(1, len(prices)):
    volatility += abs(prices[i] - prices[i - 1])

print(f"(a) Max price: ₹{max_price} on {max_day}")
print(f"(b) Min price: ₹{min_price} on {min_day}")
print(f"(c) Average:   ₹{avg:.2f}")
print(f"(d) Biggest gain: ₹{max_gain} ({gain_from} → {gain_to})")
print(f"(e) Total volatility: ₹{volatility}")
````

````python
message = input("Enter message: ")
shift = int(input("Enter shift (N): "))

encrypted = ""
for ch in message:
    if ch.isalpha():
        base = ord('A') if ch.isupper() else ord('a')
        shifted = (ord(ch) - base + shift) % 26
        encrypted += chr(base + shifted)
    else:
        encrypted += ch

print(f"Encrypted: {encrypted}")
````

### C7 — Functions (Q46–Q47)

**Q46. Write a function `calculate_gst(amount, gst_rate=18, gst_inclusive=False)`:**
- If gst_inclusive=False: amount is base, add GST on top
- If gst_inclusive=True: amount already includes GST, extract it
- Return a tuple: (base_amount, gst_amount, total_amount)
- All values rounded to 2 decimals

Test cases:
```
 calculate_gst(1000) → (1000.00, 180.00, 1180.00)
 calculate_gst(1180, gst_inclusive=True) → (1000.00, 180.00, 1180.00)
 calculate_gst(500, gst_rate=5) → (500.00, 25.00, 525.00)
```

````python
def calculate_gst(amount, gst_rate=18, gst_inclusive=False):
    if gst_inclusive:
        base = amount / (1 + gst_rate / 100)
        gst = amount - base
        total = amount
    else:
        base = amount
        gst = amount * gst_rate / 100
        total = base + gst
    return (round(base, 2), round(gst, 2), round(total, 2))
print(calculate_gst(1000))                         
print(calculate_gst(1180, gst_inclusive=True))     
print(calculate_gst(500, gst_rate=5))              
````

**Q47. Write a RECURSIVE function `power(base, exp)` that computes base raised to exp, WITHOUT
using the `**` operator and WITHOUT using `pow()`. Handle exp = 0 (return 1) and negative exp
(return 1/power(base, -exp)). Test with: power(2, 10), power(5, 0), power(3, -2).**

````python
def power(base, exp):
    """Compute base^exp recursively."""
    if exp == 0:
        return 1                         
    if exp < 0:
        return 1 / power(base, -exp)     
    return base * power(base, exp - 1)   


print(power(2, 10))
print(power(5, 0))    
print(power(3, -2))   
````

**Q49. Zomato Weekly Sales Analytics. Create a NumPy array of 7-day Zomato Bengaluru sales (in ₹):**
```
sales = np.array([12000, 18500, 14200, 22100, 19800, 35200, 28400])
 # Mon Tue Wed Thu Fri Sat Sun
```
**Without using loops, compute and print:**

(a) Total weekly sales

(b) Daily average

(c) The day (Mon/Tue/...) with maximum sales

(d) Number of days with sales above the weekly average

(e) Percentage growth from Monday to Sunday

(f) An array of day-on-day % changes (length 6)

````python
import numpy as np

sales = np.array([12000, 18500, 14200, 22100, 19800, 35200, 28400])
day_names = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

total = np.sum(sales)
print(f"(a) Total weekly sales: ₹{total:,}")

avg = np.mean(sales)
print(f"(b) Daily average:      ₹{avg:,.2f}")

max_idx = np.argmax(sales)
print(f"(c) Best day:           {day_names[max_idx]} (₹{sales[max_idx]:,})")

above_avg_count = np.sum(sales > avg)
print(f"(d) Days above average: {above_avg_count}")

pct_growth = ((sales[-1] - sales[0]) / sales[0]) * 100
print(f"(e) Growth Mon→Sun:     {pct_growth:.2f}%")

day_on_day_pct = ((sales[1:] - sales[:-1]) / sales[:-1]) * 100
print(f"(f) Day-on-day % change: {np.round(day_on_day_pct, 2)}")
````
